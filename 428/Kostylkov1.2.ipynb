```python
import numpy as np
import matplotlib.pyplot as plt
```

В массив $y$ записаны значения анодного сопротивления $R_a$ , в массив $x$ - значения практически измеренного коэффициента усиления $K_п$.

```python
y = np.array([0.5,1,3,5,10,12,15,18])
x = np.array([1.1,1.8,3,6.8,10.8,13.6,16.8,18])
```

Далее введена формула расчёта коэффициента усиления $K$ и соответствующие константы для его вычисления: сопротивление $R_{iср}$ и среднее значение статического коэффициента усиления $\mu_{ср}$.

```python
Ri = 8.67
M = 22.196
def K_t(y):
    return M/(1+Ri/y)
```

Следующий этап - черчение графиков: практического и теоретического.

```python
plt.figure(figsize=(15, 10))
#plt.plot(x, y, label='$K_{прак}$')
#plt.plot(x, K_t(y), label='$K_{теор}$')
plt.errorbar(x, y, xerr=0.3, yerr=0.8, label='$K_{прак}$')
plt.errorbar(x, K_t(y), label='$K_{теор}$')
plt.xlabel(r'$K$', fontsize=16)
plt.ylabel(r'$R_a, кОм$', fontsize=16)
plt.grid(True)
plt.legend(loc='best', fontsize=22)
plt.show()
```
