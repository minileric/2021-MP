# Найти значение функции в точках 1, 5, 10, 100

$$ y = \log_{2x}|\frac{e^{cos(x)^A}tan(x^2)}{\frac{A}{9}x +\sqrt{x^3}}| $$
Где в качестве *A* подставить день своего рождения.

```python
import matplotlib.pyplot as plt
import math
import numpy as np
def y(x,a):
    return np.log(abs(np.exp(np.cos(x)**a)*np.tan(x**2)/((a*x/9)+np.sqrt(x**3))))/np.log(2*x)

print("Значение функции в точке x = 1: ", y(1,27))
print("Значение функции в точке x = 5: ", y(5,27))
print("Значение функции в точке x = 10: ", y(10,27))
print("Значение функции в точке x = 100: ", y(100,27))
```
