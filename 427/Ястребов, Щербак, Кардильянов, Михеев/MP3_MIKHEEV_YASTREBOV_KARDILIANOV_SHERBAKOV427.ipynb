```python
from numpy import exp , log , arange
import matplotlib.pyplot as plt

def f(x,y):
    return exp(y)/(1+exp(x))

def yf(x):
    return -log(log(exp(x)+1)-x)

def print_f():
    x,y=find_y()
    plt.plot(x,y,'r',x,yf(x),'b')
    plt.show()

def find_y(eps=10**(-12)):
    y=[] ; x=[]
    h=0.001 ; x0,y0=2,2.1 ; cnt=0
    #x.append(x0) ; y.append(y0)
    a=x0+h/2 ; b=y0+h/2*f(x0,y0)
    c=f(a,b) ; delta=h*c
    y1=y0+delta ; x0+=h
    while True:
        x.append(x0) ; y.append(y1)
        y0=y1
        a=x0+h/2 ; b=y0+h/2*f(x0,y0)
        c=f(a,b) ; delta=h*c
        y1=y0+delta ; x0+=h ; cnt+=1
        if cnt>3000:
            break
        print(y1,'\t',yf(x0))
    return x,y

print_f()
```
